### Step 2: Read the data into a DataFrame


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/Users/selva/Documents/SE-25_DS/SE-25_Assignments/DS_Proj_2.1/ECA_blended_custom/TG_STAID011736.txt',
               skiprows=18, skipinitialspace=True)
df.DATE=pd.to_datetime(df.DATE.astype(str))
df
```

### Step 3: Data Wrangling
#### Create a new column by dividing the temperature by 10.
#### Plot the raw temperature data as a line plot

```python
df['new_temp'] = df.TG/10
df.plot(x='DATE',y='TG')
```

### Step 4: Time Axis

```python
df['DATE']=df.DATE.astype(str)
df.dtypes
```

```python
df['year'] = df['DATE'].str[0:4]
df
```

```python
df['month']=df['DATE'].str[5:7]
df
```

```python
df['date']=df['DATE'].str[8:10]
df
```

```python
df['new_date']=df['year']+'-'+df['month']+'-'+df['date']
df
```

```python
df1=df.set_index('new_date')
df1
```

```python
df.plot(x='new_date',y='new_temp',figsize=(10,7))
```

### Step 5: Missing Values

```python
median_val=df.new_temp.loc[df.new_temp!=-999.9].median()
df['corrected_temp']=df.new_temp.replace(-999.9,median_val)
df['corrected_temp']
```

### Step 6: Cut off the last year

```python
df=df.set_index('new_date')
df.loc[:'2021-12-31']
```

### Step 7: Aggregate

```python
ch= df.groupby('year')['corrected_temp'].median()
ch
ch.plot()
ab = pd.DataFrame(ch)
ab
```

### Step 7: Heatmap

```python
cmap = plt.get_cmap('coolwarm')
plt.figure(figsize=(15, 5))
plt.title('yearly median temperature in Berlin-Adlershof')
sns.heatmap(ab, cmap=cmap, cbar=False, yticklabels=[], xticklabels=[])
```

```python
ba = ab.transpose()
ba
```

```python
cmap = plt.get_cmap('coolwarm')

plt.figure(figsize=(15, 5))
plt.title('yearly median temperature in Berlin-Adlershof')
sns.heatmap(ba, cmap=cmap, cbar=False, yticklabels=[], xticklabels=[])
```
